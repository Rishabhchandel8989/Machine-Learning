```python
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv(r"F:\campusX\8_Pipeline\train.csv")
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
df.isnull().sum()
```

```python
# Fill in the  missing value
si_age=SimpleImputer()
X_train_age=si_age.fit_transform(X_train[['Age']])
X_test_age=si_age.fit_transform(X_test[['Age']])


si_embarked=SimpleImputer(strategy='most_frequent')
X_train_embarked=si_embarked.fit_transform(X_train[['Embarked']])
X_test_embarked=si_embarked.fit_transform(X_test[['Embarked']])
```

```python
X_train_embarked
```

```python
# onehotencoding for sex and Embarked
ohe_sex=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked=OneHotEncoder(sparse_output=False,handle_unknown='ignore')

X_train_sex=ohe_sex.fit_transform(X_train[['Sex']])
X_test_sex=ohe_sex.fit_transform(X_test[['Sex']])

X_train_embarked=ohe_embarked.fit_transform(X_train_embarked)
X_test_embarked=ohe_embarked.fit_transform(X_test_embarked)
```

```python
X_train_sex
```

```python
X_train_remaing=X_train.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test_remaing=X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_remaing
```

```python
X_train_transform=np.concatenate((X_train_remaing,X_train_age,X_train_sex,X_train_embarked),axis=1)
X_test_transform=np.concatenate((X_test_remaing,X_test_age,X_test_sex,X_test_embarked),axis=1)
```

```python
X_train_transform
```

```python
clf=DecisionTreeClassifier()
clf.fit(X_train_transform,y_train)
```

```python
# y_pred=clf.predict(X_train_transform)
y_pred = clf.predict(X_train_transform)
# accuracy_score(y_test,y_pred)
# y_pred
```

```python
# y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
